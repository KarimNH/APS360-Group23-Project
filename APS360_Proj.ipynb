```python
import mne
import matplotlib.pyplot as plt

# Load the data from an EDF file

raw1 = mne.io.read_raw_edf('/content/chb01_03.edf')
raw2 = mne.io.read_raw_edf('/content/chb01_03.edf')
#url = "https://storage.googleapis.com/chbmit-1.0.0.physionet.org/chb01/chb01_03.edf"


# Crop the data for the desired time range (in seconds)
raw1.crop(tmin=2996, tmax=3036)
raw2.crop(tmin=3100, tmax=3140)

# Get the data
data1, times1 = raw1[:]
data2, times2 = raw2[:]
# Choose the first channel
channel_1_data1 = data1[0]
channel_1_data2 = data2[0]

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(times1, channel_1_data1)
plt.title('EEG Channel 1 Displaying Seizure')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.figure(figsize=(12, 6))
plt.plot(times2, channel_1_data2)
plt.title('EEG Channel 1 No Seizure')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```
